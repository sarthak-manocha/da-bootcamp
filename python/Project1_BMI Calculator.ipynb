# **BMI Calculator**

### Body Mass Index (BMI) Calculation

The formula used to determine BMI is:

$$BMI = \frac{\text{weight in pounds} \times 703}{\text{height in inches}^2}$$

---

### BMI Classification Reference

| BMI Range | Classification | Health Risk |
| :--- | :--- | :--- |
| Under 18.5 | Underweight | Minimal |
| 18.5 - 24.9 | Normal Weight | Minimal |
| 25.0 - 29.9 | Overweight | Increased |
| 30.0 - 34.9 | Obese | High |
| 35.0 - 39.9 | Severely Obese | Very High |
| 40.0 and above | Morbidly Obese | Extremely High |

```python
name = input("Enter your name: ")
```

```python
weight = int(input("Enter your weight in pounds: "))
height = int(input("Enter your height in inches: "))
```

```python
BMI = (weight * 703) / (height * height)
print(BMI)
```

```python
if BMI>0:
    if(BMI<18.5):
        print(name +", you are underweight.")
    elif (BMI<=24.9):
        print(name +", you are f normal weight.")
    elif (BMI<29.9):
        print(name +", you are overweight.")
    elif (BMI<34.9):
        print(name +", you are obese.")
    elif (BMI<39.9):
        print(name +", you are severely obese.")
    else:
        print(name +", you are morbidly obese.")
else:
    print("Enter valid input")
```
